## Nela multi features

```python
nela_multi = pd.read_csv("processed_dataset/dataframemulti.csv")
```

```python
nela_multi.head()
```

```python
cum_avg_word_sim = []

for article_id, df_temp in nela_multi.groupby("ArticleId"):
    word_sim_sum = 0
    
    for sentence in df_temp["Fragment"]:
        prev_word = ""
        for token in nlp(sentence):
            word_sim_sum = word_sim_sum + token.similarity(nlp(prev_word))
#             print(word_sim_sum)
            prev_word = token.text
            
        if(len(nlp(sentence))>0):     
            cum_avg_word_sim.append(word_sim_sum/len(nlp(sentence)))
        else:
            cum_avg_word_sim.append(0)
#         print(cum_avg_word_sim)
#         print(len(nlp(sentence)))
```

```python
cum_avg_word_sim
```

```python
len(cum_avg_word_sim)
```

```python
nela_multi.shape
```

```python
nela_multi["avg_word_sim"] = cum_avg_word_sim
```

```python
nela_multi.head()
```

```python
cum_avg_sent_sim = []

for article_id, df_temp in nela_multi.groupby("ArticleId"):
    sent_sim_sum = 0
    prev_sentence = ""
    
    for sentence in df_temp["Fragment"]:
        sent_sim_sum = sent_sim_sum + nlp(sentence).similarity(nlp(prev_sentence))
#         print(sent_sim_sum)
        prev_sentence = sentence
        
    for count in range(len(df_temp)):
        cum_avg_sent_sim.append(sent_sim_sum/len(df_temp))
        
#     print(cum_avg_sent_sim)    
```

```python
cum_avg_sent_sim
```

```python
len(cum_avg_sent_sim)
```

```python
nela_multi["avg_sent_sim"] = cum_avg_sent_sim
```
